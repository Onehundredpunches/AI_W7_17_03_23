```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
url = 'passenger.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.hist() 
plt.show()
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr() 
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter', x = 'Miles(X)', y = 'Passenger', figsize = (9,9), color='black')
```

```python
# ham bac 1, 1 bien
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Miles(X)']), y = data['Passenger'])

print('Gia tri intercept: ', model.intercept_)  
print('Gia tri coeffient: ', model.coef_) 
print('Do chinh xac: ', model.score(X = pd.DataFrame(data['Miles(X)']), y = data['Passenger'])) 
```

```python
data.plot(kind = 'scatter', x='Miles(X)', y='Passenger', figsize=(9,9), color='black', xlim=(0,6000)) 
y_pred = model.predict(X = pd.DataFrame(data['Miles(X)'])) 
plt.plot(data['Miles(X)'], y_pred, color='red')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2]).T
model.fit(X,y = data['Passenger'])
print('Intercept la: ', model.intercept_)
print('Coefficient la:', model.coef_) 
print('Do chinh xac: ',model.score(X,y = data['Passenger']))
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2, data['Miles(X)']**3, data['Miles(X)']**4, data['Miles(X)']**5, data['Miles(X)']**6, data['Miles(X)']**7, data['Miles(X)']**8, data['Miles(X)']**9, data['Miles(X)']**10, data['Miles(X)']**11, data['Miles(X)']**12, data['Miles(X)']**13, data['Miles(X)']**14, data['Miles(X)']**15]).T
model.fit(X,y = data['Passenger'])
print('Intercept la: ', model.intercept_)
print('Coefficient la:', model.coef_)
print('Do chinh xac: ',model.score(X,y = data['Passenger']))
```

```python
range = np.arange(0, 6000, 50)
X_range = pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6, range**7, range**8, range**9, range**10, range**11, range**12, range**13, range**14, range**15]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='Miles(X)', y='Passenger', figsize=(9,9), color = 'black', xlim = (0,6000))
plt.plot(range, y_pred, color='red')
```
